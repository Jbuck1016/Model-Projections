```python
import pandas as pd

# List of CSV filenames
csv_filenames = [
    'C:\\Users\\jbuck\\Soccer Model In Python\\generating projections notebooks\\Model Projections\\la_liga_odds_5-11.csv', 
    'C:\\Users\\jbuck\\Soccer Model In Python\\generating projections notebooks\\Model Projections\\xpts_epl_odds_5-11.csv', 
    'C:\\Users\\jbuck\\Soccer Model In Python\\generating projections notebooks\\Model Projections\\serie_A_odds_5-11.csv', 
    'C:\\Users\\jbuck\\Soccer Model In Python\\generating projections notebooks\\Model Projections\\bundesliga_odds_5-11.csv', 
    'C:\\Users\\jbuck\\Soccer Model In Python\\generating projections notebooks\\Model Projections\\LIGUE_UNE_odds_5-11.csv',
    #'C:\\Users\\jbuck\\Soccer Model In Python\\generating projections notebooks\\Model Projections\\efl_odds_5-11.csv'
]

# List to store DataFrames from each CSV
dfs = []

# Iterate over each CSV filename
for filename in csv_filenames:
    # Read the CSV file into a DataFrame
    df = pd.read_csv(filename)
    # Append the DataFrame to the list
    dfs.append(df)

# Concatenate all DataFrames vertically
big_df = pd.concat(dfs, ignore_index=True)

# Display the concatenated DataFrame
print(big_df)
```

```python
import pandas as pd

def remove_trailing_zeros(value):
    """
    Function to remove trailing zeros from a decimal number represented as a string.
    It returns the value as a string without trailing zeros or as the original type.
    """
    try:
        # Convert to string and remove trailing zeros
        as_str = f"{float(value):.10g}"
        return as_str
    except ValueError:
        # Return the original value if conversion fails
        return value

# Apply the function to columns with 'Odds' in their names
for column in big_df.columns:
    if 'Odds' in column:
        big_df[column] = big_df[column].apply(remove_trailing_zeros)

# Display the DataFrame to verify the changes
print(big_df)
```

```python
# Define the file name for the CSV
csv_file_name = "big_5_leagues_odds_5-11.csv"

# Save the DataFrame to a CSV file
big_df.to_csv(csv_file_name, index=False)

print(f"DataFrame saved to {csv_file_name}")
```

```python
def color_edges(val):
    """
    Colors edges in green or red based on positive or negative values.
    """
    color = 'green' if val > 0 else 'red' if val < 0 else 'black'
    return f'color: {color}'

def format_percentage(val):
    """
    Attempts to convert a value to a float and formats it as a percentage.
    If conversion fails, returns the original value.
    """
    try:
        # Assuming your percentage values are strings like '90.90%', this will convert them to float
        val = float(val.strip('%')) / 100
        # Format as percentage with 2 decimal places
        return f'{val:.2%}'
    except ValueError:
        return val

# Applying custom styling to the DataFrame
styled_df = big_df.style.applymap(color_edges, subset=['Home Win Edge', 'Away Win Edge', 'Draw Edge'])\
                        .format(format_percentage, subset=['Home Win Probability', 'Away Win Probability', 'Draw Probability'])

# Display the styled DataFrame
styled_df
```

```python
# Example correction for deprecated applymap usage and formatting issues

# Correct approach for coloring cell backgrounds based on their values
def color_background(val):
    """
    Returns a background color style for the cell based on its value.
    """
    if val > 0:
        color = 'lightgreen'
    elif val < 0:
        color = 'salmon'
    else:
        color = 'none'
    return f'background-color: {color}'

# Apply styling to DataFrame
styled_df = big_df.style\
    .map(color_background, subset=['Home Win Edge', 'Away Win Edge', 'Draw Edge'])\
    .set_properties(**{'border': '1px solid black'})  # Adds borders to all cells

# For percentage formatting, ensure data is in a correct numeric format before applying
# Assuming 'Home Win Probability', 'Away Win Probability', and 'Draw Probability' should be formatted as percentages
for col in ['Home Win Probability', 'Away Win Probability', 'Draw Probability']:
    big_df[col] = big_df[col].str.rstrip('%').astype('float') / 100.0  # Convert from string to float
    styled_df = styled_df.format({col: '{:.2%}'})  # Apply percentage formatting

# Note: The use of map for background colors assumes a direct replacement of applymap, which is not directly available in older versions of pandas.
# If your pandas version does not support Styler.map, consider updating pandas or adapting the approach using .apply which is more broadly compatible.

styled_df
```
